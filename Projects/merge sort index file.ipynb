```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dtb
import os
from glob import glob
import datetime
import seaborn as sns
import sys

data = [pd.read_csv('C:/Users/RAPP/Documents/Capstone/HOUREMIS3_file_1-29.dat', \
                    dtype = {'SO': 'str', 'PATH': 'str', 'YEAR': 'int', 'MM': 'str', 'DD': 'str', 'HH':'str','SRC': 'str', 'RATE': 'float32', 'TEMP': 'float32' ,'VEL': 'float32'}, header = 0, delim_whitespace = True)]
#data = [pd.read_csv('C:/Users/RAPP/Documents/Capstone/HOUREMIS3_file_1-29.dat', header = 0, delimiter = " ")]

df = pd.concat(data)

```

Create NE dataframe such that operating 2013-2014

```python
NE_2yr_df = pd.DataFrame()
NE_2yr_df = df.copy()
print(NE_2yr_df['RATE'].describe())
xx = ((NE_2yr_df['YEAR'] <= 2014))
plt.plot(xx)
NE_2yr_df['RATE'][xx] = 0


#plt.plot(NE_df['RATE'])
print(NE_2yr_df['RATE'].describe())
NE_2yr_df['SRC'] = ['NE_'+ str(col) for col in NE_2yr_df['SRC'] ]



print(NE_2yr_df)
```


Create NW dataframe such that only operating 2015-2017

```python
NW_3yr_df = pd.DataFrame()
NW_3yr_df = df.copy()
print(NW_3yr_df['RATE'].describe())
#plt.plot(NW_df['RATE'])

yy = ((NW_3yr_df['YEAR'] >= 2015))
NW_3yr_df['RATE'][yy] = 0

print(NW_3yr_df['RATE'].describe())

plt.plot(NW_3yr_df['RATE'])

NW_3yr_df['SRC'] = ['NW_'+ str(col) for col in NW_3yr_df['SRC'] ]

print(NW_3yr_df)
```
