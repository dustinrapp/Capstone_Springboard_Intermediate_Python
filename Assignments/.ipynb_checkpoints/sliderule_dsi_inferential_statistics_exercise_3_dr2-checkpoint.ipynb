# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## B.1 Setup an appropriate hypothesis test.

A couple of easy tests could be performed to test the hypothesis that the correlation of the dataset is 0.  One could use the linegress function, which performs a two-sided hypothesis test, determining whether the slope is zero..


```python
from scipy.stats import linregress
linregress(clean_hospital_read_df['Number of Discharges'][81:-3], clean_hospital_read_df['Excess Readmission Ratio'][81:-3])
```

Here, we see a negative slope and a very small p value, indicating that there the slope is very likely is not zero for a linear regression fit line, even at the 1% confidence interval.  In addition, one could check the Pearson coefficient to determine how well correlated the two datasets might be:

```python
from scipy.stats.stats import pearsonr


pcoeff, p = pearsonr(clean_hospital_read_df['Number of Discharges'][81:-3], clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

print("The pearson coefficent is: " + str(pcoeff) + " and the p value is " + str(p))
```

The pearson coefficient is not near 1 or -1, so any correlation is not strong, but it there is significance in the sense that we can reject the hypothesis that the correlation is zero.   The p value is same as that predicted using the lingress function.  We can see also that there is negative correlation, indicating that as Number of Discharges increase, the Excess Readmission Ratio does decrease.


While mathematically significant, is this the trend negative enough to justify any actions?  The p value significance must be considered from from a practical standpoint as the "significant" label doesn't really tell us too much other then that there is a trend.  One thing we can say is that  the linear regression trendline suggests that for every unit decrease in number of discharges, there is a -2.86e-05 decrease in excess readmission rate. Put another way, for every decrease of 1000 discharges, there is a decrease of 0.0286 in the excess rates of admissions.

**Statistical Signficance vs Practical Significance**

Given this statistical information, should all hospitals need to demonstrate upgraded resource allocation for quality care to continue operation?  Or just a subset? Generalizing however, hospitals with a readmission rate greater then 1 probably should take action to reduce their rates.  How do we determine this threshold, though?  

A way to consider this is use a bootstrap approach and generate bootstrap replicates of slopes and begin consider confidence intervals.  

**Define a function to generate bootstrap replicate pairs of potential slopes/intercepts**

```python
def draw_bs_pairs_linreg(x, y, size=1):
    """Perform pairs bootstrap for linear regression."""

    # Set up array of indices to sample from: inds
    inds = np.arange(len(x))

    # Initialize replicates: bs_slope_reps, bs_intercept_reps
    bs_slope_reps = np.empty(size)
    bs_intercept_reps = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_inds = np.random.choice(inds, size=len(inds))
        bs_x, bs_y = x[bs_inds], y[bs_inds]
        bs_slope_reps[i], bs_intercept_reps[i] = np.polyfit(bs_x, bs_y,1)

    return bs_slope_reps, bs_intercept_reps
```

**Generate distribution of 10000 replicates**

```python
bs_slope_reps, bs_intercept_reps = draw_bs_pairs_linreg(np.array(clean_hospital_read_df['Number of Discharges'][81:-3]), np.array(clean_hospital_read_df['Excess Readmission Ratio'][81:-3]), 10000)
# Compute and print 95% CI for slope
print(np.percentile(bs_slope_reps, [2.5, 97.5]))
```

**Plot a histogram of this histogram and determine confidence intervals**

```python
conf_int = np.percentile(bs_slope_reps, [0.05, 99.5])
print("At the 1% confidence level, we can say that the slope of the trendline falls between " + str(conf_int[1]) + " and " + str(conf_int[0]))
print("Put another way, for each 1000 increases in number of discharges, the change in excess admission rates will be between " + str(conf_int[1]*1000) + " and " + str(conf_int[0]*1000))

# Plot the histogram
_ = plt.hist(bs_slope_reps, bins=50, normed=True)
_ = plt.xlabel('slope of trendline')
_ = plt.ylabel('PDF')
plt.show()
```

**A good way to visualize this is through a plot of likely trendlines..**

```python
fig, ax = plt.subplots(figsize=(16,10))

bs_lines_x= np.array([0,3000])

# Plot the bootstrap lines
for i in range(1000):
    _ = ax.plot(bs_lines_x, bs_slope_reps[i]*bs_lines_x + bs_intercept_reps[i],
                 linewidth=0.005, alpha=0.8, color='blue')

# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])


ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,700], 1.0, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([700,2500], .5, 1.0, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

While the original chart presented is the report is good for highlighting areas where where the excess of rate admissions are higher and lower, it does not provide convincing evidence supporting the claims made or the actions recommended in the report. By adding these trendlines, we can begin providing evidence that there is indeed a trend in the data suggesting that increased number of discharges does lead to a decrease in excess rate of admissions.  In addition, it provides a good visualization as to where the excess rate of readmissions drop below 1.0, which is the threshold.  The excess rate of readmissions drops below 1 (on average) when the number of discharges increases to about 700.  By plotting it this way, argument can be made more robustly that hospitals with number of discharges less then 700 may require some actions to help them decrease their excess rate of admissions.
