## Pearson correlation heatmap

**Creating a heatmap matrix showing relationships between all the features will show if some features are highly correleated, indicating collinearity.**

**Define function heatmap generation.**

```python
def heatMap(df, labels, size, method):
    #Create Correlation df
    corr = df.corr(method = method)
    #Plot figsize
    fig, ax = plt.subplots(figsize=size)
    #Generate Color Map
    colormap = sns.diverging_palette(220, 10, as_cmap=True)
    #Generate Heat Map, allow annotations and place floats in map
    
    dropSelf = np.zeros_like(corr)
    dropSelf[np.triu_indices_from(dropSelf)] = True
    sns.heatmap(corr, cmap=colormap, annot=False, fmt=".2f", mask=dropSelf)

    
    #Apply xticks
    plt.xticks(range(len(corr.columns)), corr.columns);
    #Apply yticks
    plt.yticks(range(len(corr.columns)), corr.columns)
    ax.set_xticklabels(labels, rotation = 90, ha='left')
    ax.set_yticklabels(labels, rotation = 0, ha='right')
    plt.grid(True, alpha = 0.25)
    #show plot

    
```

**Create Heatmap showing just surface feature matrix**
